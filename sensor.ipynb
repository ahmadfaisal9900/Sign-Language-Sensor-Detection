```python
# Separate features (sensor readings) and target variable (class labels)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

X = combined_df.drop('SIGN', axis=1)  # Assuming 'class_label' is your class column
y = combined_df['SIGN']
```

```python
import pandas as pd
import numpy as np

def convert_to_float(df):
  """
  Converts all data types in a DataFrame to float, handling errors and missing values,
  excluding column names.

  Args:
      df: The DataFrame to convert.

  Returns:
      A new DataFrame with all columns converted to float (where possible).
  """
  # Get original column names
  col_names = list(df.columns)

  # Convert data (excluding column names)
  try:
    df_converted = df.iloc[:,:].astype(float, errors='coerce')
  except:
    df_converted = df.copy()
    for col in df_converted.columns:
      try:
        df_converted.loc[:, col] = pd.to_numeric(df_converted[col], errors='coerce')
      except:
        df_converted.loc[:, col] = np.NAN

  # Set original column names back
  df_converted.columns = col_names

  return df_converted


X = convert_to_float(X.copy())
print(X.dtypes)
```

```python

def drop_nan_together(X, Y):
  """
  Drops rows with NaN values in X and the corresponding rows from Y.

  Args:
      X: The DataFrame containing parameters.
      Y: The DataFrame containing labels.

  Returns:
      A tuple containing the filtered DataFrames (X_filtered, Y_filtered).
  """
  # Drop rows with NaN values in X
  X_filtered = X.dropna()

  # Ensure Y has the same index as the filtered X
  Y_filtered = Y.loc[X_filtered.index]

  return X_filtered, Y_filtered

# Example usage
X_filtered, Y_filtered = drop_nan_together(X.copy(), y.copy())  # Operate on copies

print(X_filtered.shape)
print(Y_filtered.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_filtered, Y_filtered, test_size=0.4, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

```python
# # Assuming y_train_encoded and y_test_encoded are column-vectors
# y_train_encoded = y_train_encoded.ravel()
# y_test_encoded = y_test_encoded.ravel()

# # Now y_train_1d and y_test_1d are 1D arrays
# y_train_encoded.shape
```

```python
y_train_encoded = y_train_encoded.ravel()
y_test_encoded = y_test_encoded.ravel()

# Now y_train_1d and y_test_1d are 1D arrays
y_train_encoded.shape
```

```python
from sklearn.metrics import accuracy_score

kernel = 'rbf'  
svm_clf = SVC(kernel=kernel, C=0.05)  
svm_clf.fit(X_train_scaled, y_train_encoded)

y_pred_train = svm_clf.predict(X_train_scaled)
train_accuracy = accuracy_score(y_train_encoded, y_pred_train)

y_pred = svm_clf.predict(X_test_scaled)
y_pred_decoded = encoder.inverse_transform(y_pred.reshape(-1, 1))
test_accuracy = accuracy_score(y_test_encoded, y_pred)

print(f"SVM model with {kernel} kernel:")
print(f"\tTraining Accuracy: {train_accuracy:.4f}")
print(f"\tTesting Accuracy: {test_accuracy:.4f}")



```

```python
# Extract weights (w) and bias (b) from the fitted model
w = svm_clf.coef_.squeeze()  # Extract weights (assuming single output class)
b = svm_clf.intercept_

# Project weights onto the first two features for visualization (assuming 2D features)
w_normalized = w / np.linalg.norm(w)  # Normalize weights
v = w_normalized[:2]  # Project onto first two features (if more, adjust index)
```

```python
def decision_boundary(X):
  """
  This function calculates the decision boundary for the linear SVM model.
  """
  return (-b - np.dot(X, v)) / w_normalized[1]  # Assuming w_normalized[1] is for feature 2
```
